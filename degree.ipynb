```python
import argparse
import matplotlib.pylab as pl
%matplotlib inline
import networkx as nx
import my
```

```python
args = argparse.Namespace()
args.graph = 'soc-Epinions1'
# args.graph = 'soc-Slashdot0811'
# args.graph = 'soc-Slashdot0902'
```

```python
g = my.read_edgelist('data/' + args.graph)
adj = nx.adj_matrix(g)
deg = adj.sum(1)
```

```python
n, bins, patches = pl.hist(deg, bins=1000, histtype='step', log=True)
pl.xlabel('degree')
pl.ylabel('count')
pl.xscale('log')
pl.yscale('log')
```
