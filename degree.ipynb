```python
import argparse
import collections
import matplotlib.pylab as pl
%matplotlib inline
import networkx as nx
import my
```

```python
args = argparse.Namespace()
args.graph_tuple = ('soc-Epinions1', 'soc-Slashdot0811', 'soc-Slashdot0902')
```

```python
deg_dict = collections.OrderedDict()
for key in args.graph_tuple:
    g = my.read_edgelist('data/' + key)
    adj = nx.adj_matrix(g)
    deg_dict[key] = adj.sum(1)
```

```python
for key, deg in deg_dict.items():
    pl.figure()
    pl.hist(deg, bins=1000, histtype='step')
    
    pl.title(key)
    pl.xlabel('degree')
    pl.ylabel('count')
    pl.xscale('log')
    pl.yscale('log')
```
